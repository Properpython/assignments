##Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function that describes the likelihood of various outcomes occurring in an uncertain event or random experiment. In other words, it gives us the probabilities associated with different possible outcomes of the event. The sum of all probabilities in a probability distribution is equal to 1, representing the certainty that some outcome will occur.

A probability distribution provides valuable information about the possible outcomes and their associated probabilities, which allows us to make predictions and draw inferences about the event's behavior.

##Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

True Random Numbers:
True random numbers are generated from a genuinely unpredictable source, often referred to as a "random process." These sources could include physical processes in the natural world that are inherently unpredictable, such as radioactive decay, thermal noise, or chaotic systems. The key characteristic of true random numbers is that they are not reproducible, and their values cannot be predicted even if the entire history of previous random numbers is known.

Pseudo-Random Numbers:
Pseudo-random numbers, on the other hand, are generated by deterministic algorithms. These algorithms start with an initial value called the "seed" and use it to produce a sequence of numbers that appears to be random. The generated sequence is deterministic, meaning that if you know the seed value and the algorithm, you can reproduce the exact same sequence of "random" numbers. However, pseudo-random numbers are designed to have properties that make them appear random for most practical purposes.

The main reason pseudo-random numbers are considered "good enough" in many applications is that they exhibit statistical properties similar to true random numbers.

##Q3. What are the two main factors that influence the behaviour of a &quot;normal&quot; probability distribution?

The behavior of a "normal" probability distribution, also known as the Gaussian distribution or bell curve, is influenced by two main factors:

Mean (μ): The mean of a normal distribution represents its central location or the average value of the data. It is the point around which the data is symmetrically distributed. The mean determines the center of the bell curve. If the mean is shifted to the right, the entire distribution will shift to the right, and if it is shifted to the left, the distribution will shift to the left.

Standard Deviation (σ): The standard deviation of a normal distribution measures the spread or dispersion of the data points around the mean. A small standard deviation indicates that the data points are closely clustered around the mean, resulting in a narrow and tall bell curve. A large standard deviation indicates that the data points are more spread out from the mean, leading to a broader and flatter bell curve.

##Q4. Provide a real-life example of a normal distribution.

A real-life example of a normal distribution is the distribution of heights of adult people. Imagine you have a large group of adults and you measure their heights. If you create a histogram or a bar chart to display the frequency of different height ranges, you will likely observe a bell-shaped curve.

##Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term, a probability distribution may not always behave as expected due to the inherent randomness in the underlying process. In a limited number of trials, there can be significant fluctuations and deviations from the expected probabilities. However, as the number of trials grows, the behavior of the probability distribution tends to stabilize and align more closely with the expected probabilities.

Let's consider an example of flipping a fair coin, where the probability of getting a head (H) is 0.5, and the probability of getting a tail (T) is also 0.5. If we flip the coin only a few times, say 5 or 10 times, we may observe outcomes that differ from the expected probabilities. For instance, in 10 flips, we might get 7 heads and 3 tails, which deviates from the expected 5 heads and 5 tails.

##Q6. What kind of object can be shuffled by using random.shuffle?

The random.shuffle function in Python can be used to shuffle mutable sequences. Mutable sequences are sequences whose elements can be changed after they are created. The most commonly used mutable sequence types in Python are lists.

Therefore, random.shuffle can be used to shuffle elements in a list. When you pass a list to random.shuffle, it will modify the list in-place and rearrange its elements randomly, effectively shuffling the order.

##Q8. What is the relationship between exponentiation and logarithms?

Exponentiation:
Exponentiation is the operation of raising a base number to a certain power. The result of exponentiation is the value obtained when the base number is multiplied by itself a certain number of times, as determined by the exponent.

Logarithms:

Logarithms, on the other hand, are the inverse of exponentiation. The logarithm of a number to a certain base gives us the exponent to which the base must be raised to obtain the original number.


##Q9. What are the three logarithmic functions that Python supports?

Natural Logarithm (base e):

The natural logarithm function is denoted as log(x) or log(x, e), where x is the number for which the logarithm is calculated, and e is the base of the logarithm (Euler's number, approximately equal to 2.71828). If the base is not specified, the function assumes the base is e. The natural logarithm is the inverse of the exponential function with base e.

Base-10 Logarithm:

The base-10 logarithm function is denoted as log10(x). It calculates the logarithm of x with base 10. This function is used when we want to express numbers in a more human-readable way, as the base-10 logarithm gives us the number of digits before the decimal point in a given number.


Custom Base Logarithm:

Python also provides a function to calculate the logarithm with a custom base. The function is denoted as log(x, base), where x is the number for which the logarithm is calculated, and base is the desired base of the logarithm.
